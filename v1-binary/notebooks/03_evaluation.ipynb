## Step 3 — Load Data & Models

```python
# Load the test dataset
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv').squeeze()

# Load all 4 trained models from .pkl files
# pickle.load() recreates the model object exactly as it was when saved
model_files = {
    'Logistic Regression': 'logistic_regression.pkl',
    'Random Forest':       'random_forest.pkl',
    'Gradient Boosting':   'gradient_boosting.pkl',
    'XGBoost':             'xgboost.pkl'
}

models = {}
for name, fname in model_files.items():
    with open(fname, 'rb') as f:   # 'rb' = read binary mode
        models[name] = pickle.load(f)
    print(f'Loaded {name} ✅')

print(f'\nTest set: {X_test.shape[0]} students, {X_test.shape[1]} features')
```

## Step 4 — Confusion Matrices

```python
# A confusion matrix shows 4 categories of predictions:
#   True Positive  (TP) → predicted Pass, actually Pass  ✅
#   True Negative  (TN) → predicted Fail, actually Fail  ✅
#   False Positive (FP) → predicted Pass, actually Fail  ❌ (false alarm)
#   False Negative (FN) → predicted Fail, actually Pass  ❌ (missed a student)
# A good model has high TP and TN, and low FP and FN

fig, axes = plt.subplots(1, 4, figsize=(20, 4))

for ax, (name, model) in zip(axes, models.items()):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)

    # ConfusionMatrixDisplay makes a nicely formatted heatmap
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Fail', 'Pass'])
    disp.plot(ax=ax, colorbar=False, cmap='Blues')
    ax.set_title(name, fontsize=11, fontweight='bold')

plt.suptitle('Confusion Matrices — All 4 Models', fontsize=14, y=1.03)
plt.tight_layout()
plt.savefig('confusion_matrices.png', bbox_inches='tight', dpi=150)
plt.show()
print('Saved confusion_matrices.png ✅')
```

## Step 5 — ROC-AUC Curves

```python
# ROC curve plots True Positive Rate vs False Positive Rate
# at different classification thresholds
#
# AUC (Area Under the Curve):
#   AUC = 1.0 → perfect model
#   AUC = 0.5 → random guessing (the dashed diagonal line)
#   AUC > 0.8 → generally considered good
#
# We plot all 4 models on the same graph to compare them visually

plt.figure(figsize=(8, 6))
colors = ['steelblue', 'darkorange', 'green', 'red']

for (name, model), color in zip(models.items(), colors):
    # predict_proba returns probability scores instead of just 0/1 labels
    # We use the probability of class 1 (Pass) → column index [:, 1]
    y_prob = model.predict_proba(X_test)[:, 1]

    fpr, tpr, _ = roc_curve(y_test, y_prob)  # compute ROC curve points
    roc_auc = auc(fpr, tpr)                   # compute area under the curve

    plt.plot(fpr, tpr, color=color, lw=2.5, label=f'{name} (AUC = {roc_auc:.3f})')

# Diagonal line represents a random classifier (no skill)
plt.plot([0, 1], [0, 1], 'k--', lw=1.5, label='Random Baseline (AUC = 0.500)')

plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC-AUC Curves — All Models', fontsize=13, fontweight='bold')
plt.legend(loc='lower right', fontsize=10)
plt.tight_layout()
plt.savefig('roc_auc_curves.png', bbox_inches='tight', dpi=150)
plt.show()
print('Saved roc_auc_curves.png ✅')
```

## Step 6 — Model Comparison Bar Chart

```python
# Load the results summary saved by Person 2
results = pd.read_csv('results_summary.csv')

# Create a grouped bar chart with Accuracy and F1 Score side by side
x = np.arange(len(results))   # x positions for the groups
width = 0.35                   # width of each bar

fig, ax = plt.subplots(figsize=(11, 6))

bars1 = ax.bar(x - width/2, results['Accuracy'], width,
               label='Accuracy', color='steelblue', edgecolor='white')
bars2 = ax.bar(x + width/2, results['F1 Score'],  width,
               label='F1 Score',  color='darkorange', edgecolor='white')

# Add value labels on top of each bar for easier reading
for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
            f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=9, fontweight='bold')
for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
            f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=9, fontweight='bold')

ax.set_xlabel('Model', fontsize=12)
ax.set_ylabel('Score', fontsize=12)
ax.set_title('Model Comparison — Accuracy & F1 Score', fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(results['Model'], rotation=15, ha='right', fontsize=10)
ax.set_ylim(0, 1.15)
ax.legend(fontsize=11)
ax.axhline(y=0.8, color='grey', linestyle='--', alpha=0.5, label='0.8 reference line')

plt.tight_layout()
plt.savefig('model_comparison.png', bbox_inches='tight', dpi=150)
plt.show()
print('Saved model_comparison.png ✅')
```

## Step 7 — SHAP Feature Importance (XGBoost)

```python
# SHAP (SHapley Additive exPlanations) explains WHY the model makes each prediction
# It assigns each feature an 'importance score' based on how much it changes the output
#
# This is the most impressive part of the project from a presentation standpoint!
# It answers: 'Which student features matter most for predicting pass/fail?'
#
# We use it on XGBoost since TreeExplainer is fast and accurate for tree-based models

xgb_model = models['XGBoost']

# TreeExplainer is specifically designed for tree-based models (RF, XGBoost, etc.)
# It's much faster than the generic KernelExplainer
explainer = shap.TreeExplainer(xgb_model)

# Calculate SHAP values for each prediction in the test set
# shap_values[i][j] = how much feature j pushed prediction i towards class 1
shap_values = explainer.shap_values(X_test)

# Bar plot: average absolute SHAP value per feature = overall feature importance
plt.figure(figsize=(10, 7))
shap.summary_plot(
    shap_values,
    X_test,
    plot_type='bar',   # bar chart of mean importance
    show=False
)
plt.title('SHAP Feature Importance — XGBoost\n(Higher = more influential in predictions)',
          fontsize=12)
plt.tight_layout()
plt.savefig('shap_importance_bar.png', bbox_inches='tight', dpi=150)
plt.show()
print('Saved shap_importance_bar.png ✅')
```

## Step 8 — SHAP Beeswarm Plot (Detailed Impact)

```python
# The beeswarm (dot) plot shows MORE information than the bar chart:
#   - Each dot = one student in the test set
#   - Dot position on X axis = SHAP value (positive = pushed towards Pass)
#   - Dot COLOR = actual feature value (red = high, blue = low)
#
# Example interpretation: if 'failures' is red (high failures) and far left (negative SHAP)
# → having many failures strongly pushes prediction towards Fail

plt.figure(figsize=(10, 7))
shap.summary_plot(
    shap_values,
    X_test,
    show=False   # dot/beeswarm plot (default when plot_type not specified)
)
plt.title('SHAP Beeswarm — Feature Impact Direction & Magnitude',
          fontsize=12)
plt.tight_layout()
plt.savefig('shap_beeswarm.png', bbox_inches='tight', dpi=150)
plt.show()
print('Saved shap_beeswarm.png ✅')
```
