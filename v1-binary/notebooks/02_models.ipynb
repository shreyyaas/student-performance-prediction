## Step 1 — Install & Import Libraries

```python
# Install XGBoost — a powerful gradient boosting library not in default Colab
!pip install xgboost -q

import pandas as pd
import numpy as np
import pickle          # for saving trained models to disk
import matplotlib.pyplot as plt
import seaborn as sns

# sklearn models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# sklearn utilities
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, classification_report

sns.set_theme(style='whitegrid')
print('All libraries imported ✅')
```

## Step 2 — Upload & Load Processed Data

```python
# Upload the 4 CSV files that Person 1 saved
# Click the 'Choose Files' button that appears and select all 4 files
from google.colab import files
print('Please upload: X_train.csv, X_test.csv, y_train.csv, y_test.csv')
uploaded = files.upload()
```

```python
# Load the uploaded files into DataFrames
X_train = pd.read_csv('X_train.csv')
X_test  = pd.read_csv('X_test.csv')

# .squeeze() converts a single-column DataFrame into a pandas Series
# (sklearn models expect a 1D array for the target, not a 2D DataFrame)
y_train = pd.read_csv('y_train.csv').squeeze()
y_test  = pd.read_csv('y_test.csv').squeeze()

print(f'X_train: {X_train.shape} — {X_train.shape[0]} students, {X_train.shape[1]} features')
print(f'X_test : {X_test.shape}')
print(f'y_train: {y_train.shape} — class balance: {y_train.value_counts().to_dict()}')
print(f'y_test : {y_test.shape}')
```

## Step 3 — Model 1: Logistic Regression (Baseline)

```python
# Logistic Regression is our baseline model — the simplest possible classifier
# It finds a linear boundary between pass and fail students
# We use this as a benchmark: all other models should beat this
#
# max_iter=1000  → allow more iterations for the optimizer to converge
# random_state=42 → for reproducibility

print('Training Logistic Regression...')
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)          # train the model on training data
y_pred_lr = lr.predict(X_test)    # make predictions on test data

# cross_val_score gives a more reliable accuracy estimate
# It splits training data into 5 folds and trains/tests 5 times
cv_lr = cross_val_score(lr, X_train, y_train, cv=5, scoring='f1')

print('\n=== Logistic Regression Results ===')
print(f'Test Accuracy   : {accuracy_score(y_test, y_pred_lr):.4f}')
print(f'Test F1 Score   : {f1_score(y_test, y_pred_lr):.4f}')
print(f'CV F1 (5-fold)  : {cv_lr.mean():.4f} ± {cv_lr.std():.4f}')
print('\nDetailed Report:')
print(classification_report(y_test, y_pred_lr, target_names=['Fail', 'Pass']))
```

## Step 4 — Model 2: Random Forest + GridSearchCV Tuning

```python
# Random Forest builds many decision trees and combines their predictions
# It's more powerful than a single tree because the trees vote together
# This reduces overfitting (memorizing training data instead of learning patterns)
#
# GridSearchCV tries every combination of hyperparameters below
# and picks the combination with the best cross-validation F1 score
#
# Hyperparameters being tuned:
#   n_estimators    → how many trees to build (more = better but slower)
#   max_depth       → how deep each tree can grow (None = unlimited)
#   min_samples_split → minimum samples needed to split a node

rf_params = {
    'n_estimators':     [100, 200],
    'max_depth':        [None, 5, 10],
    'min_samples_split': [2, 5]
}

print('Training Random Forest with GridSearchCV (this may take ~1 minute)...')
rf = RandomForestClassifier(random_state=42)
rf_grid = GridSearchCV(
    rf,
    rf_params,
    cv=5,            # 5-fold cross validation for each param combination
    scoring='f1',    # optimize for F1 score, not just accuracy
    n_jobs=-1,       # use all CPU cores to speed things up
    verbose=1        # print progress
)
rf_grid.fit(X_train, y_train)

best_rf = rf_grid.best_estimator_    # the model with best hyperparameters
y_pred_rf = best_rf.predict(X_test)

print(f'\nBest hyperparameters found: {rf_grid.best_params_}')
print('\n=== Random Forest Results ===')
print(f'Test Accuracy : {accuracy_score(y_test, y_pred_rf):.4f}')
print(f'Test F1 Score : {f1_score(y_test, y_pred_rf):.4f}')
print('\nDetailed Report:')
print(classification_report(y_test, y_pred_rf, target_names=['Fail', 'Pass']))
```

## Step 5 — Model 3: Gradient Boosting

```python
# Gradient Boosting also builds multiple trees, but differently from Random Forest:
# It builds trees SEQUENTIALLY — each new tree corrects the mistakes of the previous one
# This 'boosting' process makes it very accurate
#
# Key hyperparameters:
#   n_estimators  → number of boosting rounds (trees)
#   learning_rate → how much each tree contributes (smaller = more conservative)
#   max_depth     → max depth of each individual tree

print('Training Gradient Boosting...')
gb = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=4,
    random_state=42
)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)

print('\n=== Gradient Boosting Results ===')
print(f'Test Accuracy : {accuracy_score(y_test, y_pred_gb):.4f}')
print(f'Test F1 Score : {f1_score(y_test, y_pred_gb):.4f}')
print('\nDetailed Report:')
print(classification_report(y_test, y_pred_gb, target_names=['Fail', 'Pass']))
```

## Step 6 — Model 4: XGBoost + GridSearchCV Tuning

```python
# XGBoost (Extreme Gradient Boosting) is a highly optimized version of Gradient Boosting
# It's one of the most popular algorithms for structured/tabular data
# Used in many Kaggle competition winning solutions
#
# Additional tuning vs standard Gradient Boosting:
#   learning_rate → we try multiple values (0.05, 0.1, 0.2)
#   max_depth     → controls model complexity

xgb_params = {
    'n_estimators':  [100, 200],
    'max_depth':     [3, 5, 7],
    'learning_rate': [0.05, 0.1, 0.2]
}

print('Training XGBoost with GridSearchCV (this may take ~2 minutes)...')
xgb = XGBClassifier(
    random_state=42,
    eval_metric='logloss',     # evaluation metric during training
    use_label_encoder=False,   # suppress a deprecation warning
    verbosity=0
)
xgb_grid = GridSearchCV(
    xgb, xgb_params, cv=5, scoring='f1', n_jobs=-1, verbose=1
)
xgb_grid.fit(X_train, y_train)

best_xgb = xgb_grid.best_estimator_
y_pred_xgb = best_xgb.predict(X_test)

print(f'\nBest hyperparameters found: {xgb_grid.best_params_}')
print('\n=== XGBoost Results ===')
print(f'Test Accuracy : {accuracy_score(y_test, y_pred_xgb):.4f}')
print(f'Test F1 Score : {f1_score(y_test, y_pred_xgb):.4f}')
print('\nDetailed Report:')
print(classification_report(y_test, y_pred_xgb, target_names=['Fail', 'Pass']))
```

## Step 7 — Results Comparison Table

```python
# Summarize all model results in one table for easy comparison
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'Gradient Boosting', 'XGBoost'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_rf),
        accuracy_score(y_test, y_pred_gb),
        accuracy_score(y_test, y_pred_xgb)
    ],
    'F1 Score': [
        f1_score(y_test, y_pred_lr),
        f1_score(y_test, y_pred_rf),
        f1_score(y_test, y_pred_gb),
        f1_score(y_test, y_pred_xgb)
    ]
})

# Sort by F1 Score descending — best model at top
# We use F1 over accuracy because it accounts for both precision and recall
results = results.sort_values('F1 Score', ascending=False).reset_index(drop=True)

print('='*45)
print('        MODEL COMPARISON RESULTS')
print('='*45)
print(results.to_string(index=False))
print('='*45)
print(f'\n🏆 Best Model: {results.iloc[0]["Model"]}')

# Save results for Person 3 to use in the comparison chart
results.to_csv('results_summary.csv', index=False)
print('Results saved to results_summary.csv ✅')
```

## Step 8 — Save All Trained Models

```python
# pickle saves Python objects (including trained ML models) to a binary file
# Person 3 will load these .pkl files to run evaluation without retraining

models_to_save = {
    'logistic_regression': lr,
    'random_forest':       best_rf,
    'gradient_boosting':   gb,
    'xgboost':             best_xgb
}

for name, model in models_to_save.items():
    filename = f'{name}.pkl'
    with open(filename, 'wb') as f:   # 'wb' = write binary mode
        pickle.dump(model, f)
    print(f'Saved {filename} ✅')

print('\nDownload these .pkl files + results_summary.csv and share with Person 3')
print('Download from: Files panel (folder icon) on the left in Colab')
```
